```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
#Exploring dataset to fill in NaN's of population column in our main dataset (Life_expectancy_data)
# Load the dataset into the notebook (from United Nations)

UN_population_path = "Resources/UNdata_popdata.csv"
```

```python
# create the dataframe without the numbered rows (can use index = 0 to make country column index if needed)

UN_population_df = pd.read_csv(UN_population_path)
print(UN_population_df.shape)
UN_population_df.head()
```

```python
#Identify NaN's by column
for column in UN_population_df.columns:
    print (f"Column {column} has {UN_population_df[column].isnull().sum()} null values")
```

```python
#Identify counts by column where value = 0
for column in UN_population_df.columns:
    count = (UN_population_df[column] == 0).sum()
    print(f"Count of zeros in Column {column} : ", count)
```

```python
#Dataset needs to be slimmed down for a better merging to 1st dataset
#Drop years that are before 2000 and after 2015
UN_population_df.drop(UN_population_df[UN_population_df.Year < 2000].index, inplace = True)
UN_population_df.drop(UN_population_df[UN_population_df.Year > 2015].index, inplace = True)

#Check shape
print(UN_population_df.shape)
UN_population_df.head()
```

```python
#Drop more unnecessary rows for our purpose 
# where sex equals male or female and area equals urban or rural
UN_population_df.drop(UN_population_df[UN_population_df.Area == "Urban"].index, inplace = True)
UN_population_df.drop(UN_population_df[UN_population_df.Area == "Rural"].index, inplace = True)
UN_population_df.drop(UN_population_df[UN_population_df.Sex == "Male"].index, inplace = True)
UN_population_df.drop(UN_population_df[UN_population_df.Sex == "Female"].index, inplace = True)

#Check shape and head
print(UN_population_df.shape)
UN_population_df.head(50)
```

### Some countries have two rows for the same year such as Albania, 2015.  The numbers are different because of the source year.  We will assume the more recent source year is more accurate and drop the 2nd instance.

```python
#Rename column so it's the same as the Life_Expectancy_data
UN_population_df.rename(columns = {"Country or Area": "Country"}, inplace = True)
```

```python
#Drop rows where the Year is not unique for the country and assign it to a new DF
filtered_UN_df = UN_population_df.drop_duplicates(["Country", "Year"])
filtered_UN_df.head(50)
```

```python
#Look at number of countries after slimming down to compare to the 193 countries in Life_expectancy_data
print(filtered_UN_df["Country"].nunique())
filtered_UN_df.Country.unique()
```

```python
#Drop unnecessary columns for our purpose
UN_cleaned_df = filtered_UN_df.drop(columns = ["Area", "Sex", "Record Type", "Reliability", "Source Year", "Value Footnotes"], axis = 1)
```

```python
#Rename column value and inspect dataframe

UN_cleaned_df = UN_cleaned_df.rename(columns={"Value" : "UN_population"})
UN_cleaned_df.head(50)
```
