```python
#Import dependencies
import pandas as pd
```

```python
#Load the dataset into the notebook
file_path = "Resources/GDP_by_Country.csv"
```

```python
#Create the dataframe
gdp_df = pd.read_csv(file_path)
print(gdp_df.shape)
gdp_df.head()
```

```python
#drop excess columns and convert index
df_gdp_wide = gdp_df.drop(gdp_df.columns[[1,2,3]], axis=1).set_index(["Country"])
df_gdp_wide.head()
```

```python
#convert df from wide to long
year_list = list(df_gdp_wide.columns)
df_gdp_long = pd.melt(df_gdp_wide, value_vars = year_list, 
                      value_name = "GDP2", var_name = "Year", ignore_index = False)
df_gdp_long.head(20)
```

```python
df_gdp_long.dtypes
```

```python
#Convert Year to INT

df_gdp_long["Year"] = pd.to_numeric(df_gdp_long["Year"])
df_gdp_long["GDP2"] = [float(str(i).replace(",", "")) for i in df_gdp_long["GDP2"]]
```

```python
df_gdp_long.dtypes
```

```python
# Drop all rows outside of 2000-2015
df_gdp_2000_2015 = df_gdp_long.loc[
    (df_gdp_long['Year']>1999) & (df_gdp_long['Year']<2016)].sort_values(by = ['Country','Year'])
print(df_gdp_2000_2015.shape)
df_gdp_2000_2015.head(20)
```

```python
#Count Countries and NANs
print(df_gdp_2000_2015['GDP2'].isna().sum())

len(df_gdp_2000_2015.index.unique())
```

```python
df_gdp_2000_2015.to_csv("Resources/GDP_cleaned.csv")
```
