```python
# Initial imports
import pandas as pd
import hvplot.pandas
from pathlib import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Load the the dataset into the notebook

file_path = "C:/Users/theod/Bootcamp/FINAL_PROJECT/Module_20_Project/Data/CO2_emission 1990-2019.csv"
```

```python
# create the dataframe without the numbered rows by using index_col=0
df = pd.read_csv(file_path,index_col=0)
df.head()
```

```python
# examine how the data is shaped
# this is help to understand if we are dealing with a long or wide set

print(df.shape)
```

```python
# Check out the data types: this could reveal additonal conversions that might need to happen 

df.dtypes
```

```python
# Change index name
df.index.names = ['Country']
```

```python
# Check on the data
print(df.shape)
df.head(5)
```

```python
# Remove rows that have at least 1 null value.

df.dropna(how='any', inplace=True)

# Check the dataset shape and head to make sure the nulls were removed

print(df.shape)

df.head()
```

```python
# Drop Columns 
df.drop(axis=1, labels=["country_code", "Region", "Indicator Name"], inplace=True)
df.head()
```

```python
df.drop([col for col in df.columns if '20' not in col],axis=1,inplace=True)
```

```python
df.drop(['2016', '2017', '2018', '2019', '2019.1'], axis=1, inplace=True)
df.head()
```

```python
df = df.reset_index()
df = pd.melt(df, id_vars='Country')
```

```python
df = df.rename(columns={'variable': 'Year', 'value': 'Emissions'})
```

```python
df = df.set_index('Country')
```

```python
print(df.shape)
df.head()
```

```python
# Export to csv
df.to_csv("emissions.csv")
```
